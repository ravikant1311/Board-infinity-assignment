```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('company_sales_data.csv')
```

```python
df.head()
```

# Extract the product names into a list from the data frame

```python
productName=list(df.columns)
productName[1:-2]
```

# Create a two-dimensional list of product names and sales

```python
product=[]
Sales=[]
for feature in df:
    Total_Sales=df[feature].sum()
    product.append(feature)
    Sales.append(Total_Sales)
```

```python
[product,Sales]
```

# Create a dictionary of sold items

```python
for i, k in zip(product,Sales):
    a={i:k}
    print(a)
```

# What is the average profit of the company in a month

```python
df['total_profit'].mean()
```

# Get the most sold item of the company

```python
sales_product=pd.DataFrame([product,Sales])
print(sales_product)
```

```python
df1=sales_product.transpose()
```

# Read Total profit of all months and show it using a line plot

```python
plt.figure(figsize = (10,7))
plt.plot(df.month_number,df.total_profit,color= 'r')
plt.xlabel('Months')
plt.ylabel('Profit')
plt.title('Profit Analysis')
plt.show();
```

# Read all product sales data and show it using a Multiline Plot

```python
plt.figure(figsize = (16,9))
plt.plot(df.month_number,df.facecream, label='facecream')
plt.plot(df.month_number,df.shampoo,label='shampoo')
plt.plot(df.month_number,df.facewash,label='facewash')
plt.plot(df.month_number,df.bathingsoap,label='bathingsoap')
plt.plot(df.month_number,df.moisturizer,label='moisturizer')
plt.xlabel('Months')
plt.ylabel('Total Sales')
plt.title('Sales Analysis')
plt.show();
```

# Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.figure(figsize = (9,5))
plt.scatter(df.month_number, df.toothpaste,color = 'r')
plt.xlabel('Month Number')
plt.ylabel('Total Sales')
plt.title('Sales Analysis')
plt.show();
```

# Read face cream and face wash product sales data and show it using the bar chart

```python
plt.figure(figsize = (16,9))
df.plot(x="month_number", y=["facecream", "facewash"], kind="bar")
plt.xlabel('Month Number')
plt.ylabel('Total Sales')
plt.title('Sales Analysis of Facecream and Facewash')
plt.show();
```

# Read Bathing soap and face wash data of all months and display it using a Subplot

```python
fig = plt.figure(figsize =(13, 10))
plt.subplot(2,2,1)
plt.plot(df.month_number, df.bathingsoap)
plt.xlabel('Month Number')
plt.ylabel('Total Sales')
plt.title('Sales Analysis for bathingsoap')
plt.subplot(2,2,2)
plt.plot(df.month_number,df.facewash)
plt.xlabel('Month Number')
plt.ylabel('Total Sales for facewash')
plt.title('Sales Analysis')
plt.show();
```
